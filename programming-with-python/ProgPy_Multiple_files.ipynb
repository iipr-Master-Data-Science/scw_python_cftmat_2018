# Multiple files

We almost have all the tools needed to process all files. We need an extra library that will make the task easier:

```python
import glob
```

Glob has a function called <code>glob</code> that finds files and directories whose names match a pattern:

```python
print(glob.glob('inflammation-*.csv'))
```

```python
for file in glob.glob('./*.csv'):
    print(file)
```

Output is an unsorted list of filenames to loop over. We can sort it using <code>sorted</code> built-in

```python
for file in sorted(glob.glob('./inflammation-*.csv')):
    print(file)
```

Now we can bring together all loops, lists and loops over lists, and use it to process all files: 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = sorted(glob.glob('./inflammation-*.csv'))[:3]
for file in files:
    data = np.loadtxt(file, delimiter=',')
    fig, axarr = plt.subplots(1, 3)
    fig.set_size_inches(10, 4)
    
    axarr[0].plot(data.mean(axis=0))
    axarr[0].set_ylabel('mean', fontsize=15)
    
    axarr[1].plot(data.min(axis=0))
    axarr[1].set_ylabel('min', fontsize=15)
    
    axarr[2].plot(data.max(axis=0))
    axarr[2].set_ylabel('max', fontsize=15)
    
    plt.tight_layout()   
    plt.show()
```

### Plotting Differences

Plot the difference between the average of the first dataset and the average of the second dataset, i.e., the difference between the leftmost plot of the first two figures.

```python
data_0 = np.loadtxt('inflammation-01.csv', delimiter=',')
data_1 = np.loadtxt('inflammation-02.csv', delimiter=',')

fig, ax = plt.subplots()
fig.set_size_inches(10, 4)
ax.plot(data_0.mean(axis=0) - data_1.mean(axis=0))
ax.set_ylabel('difference', fontsize=15)
plt.tight_layout()   
plt.show()
```
